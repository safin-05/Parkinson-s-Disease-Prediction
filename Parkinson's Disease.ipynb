```python
# Finding the distribution of the dataset
## Below function is used to visualize the hisplot and KDE(probability curve)

def displot(col):
  sns.displot(df[col],kde = True)
  plt.show()

for i in list(df.columns)[1:]:
  displot(i)
```

```python
## From the above the graphs we can determine that most of the features are right tailed
## Meaning there are outliers
```

```python
## Finding outliers using boxplot
def boxplot(col):
  sns.boxplot(df[col])
  plt.show()

for i in list(df.columns)[1:]:
  boxplot(i)

## Since this is a medical dataset, it is advisible to not treat outliers
## To create a ML model use feature scaling like MinMaxScaler to handle outliers
```

```python
## Analysing correlation between variables
plt.figure(figsize = (20,20))
corr = df.corr()
sns.heatmap(corr, annot = True, cmap = 'rainbow')
```

```python
## Most of the data are heavily correlated
## But our dependant variable is not highly related to any independant variable
## However multi collinearity is observed (features having same values are multi collinear)
```

```python
# Creating X and y for train-test split
X = df.drop(['status', 'name'], axis= 1)
y = df['status']
```

```python
X.head()
```

```python
## Oversampling data
from imblearn.over_sampling import RandomOverSampler
ros = RandomOverSampler()
X_ros, y_ros = ros.fit_resample(X,y)
print(y.value_counts())
print('###############'*10)
print(y_ros.value_counts())
```

```python
# Feature Scaling
from sklearn.preprocessing import MinMaxScaler
```

```python
scalar = MinMaxScaler((-1,1))
X = scalar.fit_transform(X_ros)
y = y_ros
```
